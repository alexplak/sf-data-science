## Задание 1.4

```python
from sympy import symbols, solve, Eq

x, y, l1 = symbols("x, y, l1")
f = x**2 + 2*y**2
L = x**2 + 2*y**2 + l1*(x + y - 20)

dL_dx = L.diff(x)
dL_dy = L.diff(y)
dL_dl1 = L.diff(l1)

opt_arg = solve([dL_dx, dL_dy, dL_dl1], [x, y, l1])
display(opt_arg)
f.evalf(subs={x: opt_arg[x], y: opt_arg[y]})
```

## Задание 2.7

```python
import seaborn as sns
import pandas as pd
import ssl
import numpy as np

ssl._create_default_https_context = ssl._create_unverified_context

df = sns.load_dataset('diamonds')

display(df)
```

```python
df.drop(['depth', 'table', 'x', 'y', 'z'], axis=1, inplace=True)

df = pd.get_dummies(df, drop_first=True)

display(df)
```

```python
df['carat'] = np.log(1+df['carat'])
df['price'] = np.log(1+df['price'])

display(df)
```

```python
X = df.drop(columns="price")
y = df["price"]
```

```python
from sklearn import model_selection
X_train, X_valid, y_train, y_valid = model_selection.train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn import linear_model, metrics

param_grid = {
    "loss": ["squared_error", "epsilon_insensitive"],
    "penalty": ["elasticnet"],
    "alpha": np.logspace(-3, 3, 10),
    "l1_ratio": np.linspace(0, 1, 10),
    "learning_rate": ["constant"],
    "eta0": np.logspace(-4, -1, 4)
}

grid_search = model_selection.GridSearchCV(
    estimator=linear_model.SGDRegressor(random_state=42), 
    param_grid=param_grid,
    #scoring=metrics.mean_squared_error,
    n_jobs = -1
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
#print(grid_search.best_score_)
y_train_pred = grid_search.predict(X_train)
print(metrics.mean_squared_error(y_train, y_train_pred))
```

## Задание 4.1

```python
from scipy.optimize import minimize
import numpy as np

def func(x):
    return x[0]**2 - x[0]*x[1] + x[1]**2 + 9*x[0] - 6*x[1] + 20

def grad_func(x):
    return np.array([2*x[0] - x[1] + 9, -x[0] + 2*x[1] - 6])

x_0 = [-400, -400]
result = minimize(func, x_0, method = 'BFGS', jac=grad_func)

print('Статус оптимизации %s' % result['message'])
print('Количество оценок: %d' % result['nfev'])
solution = result['x']
evaluation = func(solution)
print('Решение: f(%s) = %.5f' % (solution, evaluation))
```
