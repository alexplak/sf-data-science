```python
df_train = pd.read_csv('./data/train_data.csv')
df_test = pd.read_csv('./data/test_data.csv')
df_submission = pd.read_csv('./data/sample_submission.csv')

display(df_train.shape)
display(df_test.shape)
display(df_submission.shape)
```

```python
display(df_train.describe())
display(df_train.info())
```

```python
display(df_test.describe())
display(df_test.info())
```

```python
df_submission['final_price'] = df_test['vehicle_year'].apply(lambda x: df_train[df_train['vehicle_year'] == x]['final_price'].median() if (df_train[df_train['vehicle_year'] == x].shape[0] != 0) else df_train['final_price'].median())
display(df_submission, df_submission.shape)
```

```python
df_submission['final_price'] = df_test.apply(
    lambda x: df_train[(df_train['vehicle_year'] == x['vehicle_year']) & (df_train['vehicle_manufacturer'] == x['vehicle_manufacturer'])]['final_price'].median()
        if (df_train[(df_train['vehicle_year'] == x['vehicle_year']) & (df_train['vehicle_manufacturer'] == x['vehicle_manufacturer'])].shape[0] != 0)
        else df_train['final_price'].median(),
        axis=1)
```
